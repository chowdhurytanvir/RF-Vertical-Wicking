```python
f = open('Vertical_Wicking.txt')
f.readline()  # skip the header
data = np.loadtxt(f)
```

```python
#from numpy import genfromtxt
#data = genfromtxt('Vertical_Wicking.csv', delimiter=',')
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
```

```python
X = data[:, 1:]  # select columns 1 through end
y = data[:, 0]
```

```python
training_size=0.3
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1-training_size,random_state=42)
```

```python
print (X_train.shape, y_train.shape)
print (X_test.shape, y_test.shape)
```

```python
forest_model = RandomForestRegressor(random_state=2)
forest_model.fit(X_train, y_train)
melb_preds = forest_model.predict(X_test)
print(mean_absolute_error(y_test, melb_preds))
```

```python
forest_model.score(X_test, y_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, melb_preds)
```

```python
plt.scatter(y_test, melb_preds)
plt.axis([0.0015, .0085, 0.0015, .0085])
plt.xlabel('Wicking Front (m)',fontsize=14)
plt.ylabel('Predicted Wicking Front (m)',fontsize=14) 
plt.savefig('Training Size 40%.jpg',dpi=1000) ## svg for high quality. Change dpi for higher quality
```
